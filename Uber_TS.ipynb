```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading dataset
df = pd.read_csv(r"C:\Users\Admin\Documents\GitHub\Uber-time-series-analysis\trips_data-0.csv")
```

```python
# Checking the data type of columns
df.info()
```

```python
# Converting the data type and creating a new column for the hours only
df['request_time'] = pd.to_datetime(df['request_time'])
df['hour'] = df['request_time'].dt.hour
```

```python
df.info()
```

```python
df.head()
```

```python
# Group rides by hour of the day and ride status
# - Counts how many rides occurred for each status (completed, canceled, etc.) in each hour
# - Converts the result into a table where each status is a separate column
# - Missing combinations are filled with 0
hourly_counts = df.groupby(['hour', 'status']).size().unstack(fill_value=0)

# Show the number of rides by status for each hour
hourly_counts
```

```python
# Calculate total cancellations per hour
# - Sum across all cancellation categories: driver_canceled, rider_canceled, and failed
# - Store the result in a new column 'canceled_total'
hourly_counts['canceled_total'] = hourly_counts[['driver_canceled', 'rider_canceled', 'failed']].sum(axis=1)

# Display the updated table with the new 'canceled_total' column
hourly_counts
```

```python
# Calculate total rides per hour
# - Add completed rides and all cancellations together
# - This avoids double-counting, since 'canceled_total' already combines all cancellation types
hourly_counts['total'] = hourly_counts['completed'] + hourly_counts['canceled_total']

# Display the updated table with the new 'total' column
hourly_counts
```

```python
# Calculate cancellation rate per hour
# - Divide total cancellations by total rides
# - Gives the proportion of rides that were canceled for each hour (value between 0 and 1)
hourly_counts['cancel_rate'] = hourly_counts['canceled_total'] / hourly_counts['total']

# Display the updated table with the new 'cancel_rate' column
hourly_counts
```

```python
# Plot bar chart: total number of cancellations per hour
plt.figure(figsize=(10,5))

# Bar height = number of cancellations; X-axis = hours of the day
plt.bar(hourly_counts.index, hourly_counts['canceled_total'], color='red')

# Add labels and title
plt.xlabel("Hour of Day")
plt.ylabel("Number of Cancellations")
plt.title("Uber Ride Cancellations by Hour of Day")

# Show all 24 hours on the x-axis
plt.xticks(range(0,24))

# Add horizontal grid lines for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.show()
```

```python
# Plot line chart: cancellation rate (proportion of rides canceled) per hour
plt.figure(figsize=(10,5))

# Plot cancellation rate by hour with markers for each point
plt.plot(hourly_counts.index, hourly_counts['cancel_rate'], marker='o', color='blue')

# Add labels and title
plt.xlabel("Hour of Day")
plt.ylabel("Cancellation Rate")
plt.title("Uber Ride Cancellation Rate by Hour of Day")

# Show all 24 hours on the x-axis
plt.xticks(range(0,24))

# Add grid lines for easier interpretation
plt.grid(True, linestyle='--', alpha=0.7)

# Display the chart
plt.show()
```

```python
# Plot combined chart: cancellations (bar) + cancellation rate (line) per hour
fig, ax1 = plt.subplots(figsize=(10,5))

# Bar chart for absolute number of cancellations (left y-axis, red)
ax1.bar(hourly_counts.index, hourly_counts['canceled_total'], color='red', alpha=0.6)
ax1.set_xlabel("Hour of Day")
ax1.set_ylabel("Number of Cancellations", color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Line plot for cancellation rate (right y-axis, blue)
ax2 = ax1.twinx()
ax2.plot(hourly_counts.index, hourly_counts['cancel_rate'], marker='o', color='blue')
ax2.set_ylabel("Cancellation Rate", color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Add title, x-axis ticks, and grid
plt.title("Uber Ride Cancellations: Count vs Rate by Hour of Day")
plt.xticks(range(0,24))
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the chart
plt.show()
```
